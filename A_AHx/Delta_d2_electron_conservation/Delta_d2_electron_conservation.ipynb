```python
import numpy as np
from sklearn.model_selection import train_test_split
import sys
import os
from catGP import preprocess_data, OMGP
from catSC import Linear_Scaling
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
'''data_format = [metadata, slab_d0, slab_d1, bimetal_fp, 
                  ads_d0_unscaled, ads_d1_unscaled, ads_d2_unscaled, SS_unscaled,
                  ads_d0_scaled, ads_d1_scaled, ads_d2_scaled, SS_scaled,
                  energy, energy_A]'''


data_scaling = np.load('data_delta_GP.npy')[()]
data_ml = np.load('CH3_data.npy')
y = np.array([i[1][0] for i in data_scaling])
y_sc = np.array([i[0][0] for i in data_scaling])
md = [i[3] for i in data_scaling]
del data_scaling

X = []
for mmdd in md:
    des1 = (mmdd[1], mmdd[2], mmdd[3], mmdd[4])
    print(des1)
    for dml in data_ml:
        if dml[0] == des1:
            X += [np.hstack([dml[1].reshape(-1), 
                             dml[2].reshape(-1),
                             dml[3].reshape(-1), 
                             dml[4].reshape(-1),
                             dml[5].reshape(-1),
                             dml[6].reshape(-1),
                             dml[7].reshape(-1),
                             dml[8].reshape(-1),
                             dml[9].reshape(-1),
                             dml[10].reshape(-1),
                             dml[11].reshape(-1)])] 

X = np.vstack(X)
    
```

```python
data = preprocess_data(X, y)
data.clean_data()
X, y = data.get_data()
assert len(y) == len(y_sc)
    
sc = Linear_Scaling(y_sc, y, 'CH', 'CH3', 'eV')
sc.get_coeff()
a = sc.plot_scaling()

scaling_dict = sc.__dict__
print('Scaling relation: slope = {} eV\nintercept = {}eV'.format(sc.slope,
                                                                 sc.intercept))

kernel_recipe = {'ConstantKernel' : [{'RBF' : [1.0,
                                  {'length_scale' : 1.0}]},
                                  {'constant_value' : 1.0,
                                  'constant_value_bounds' : (3e-7, 3e7)}],
                 'WhiteKernel' : {'noise_level' : 0.1,
                                  'noise_level_bounds' : (1e-5, 1e5)}}

r_state = [10, 20, 42, 80]
r_data = {rs : {} for rs in r_state}
for rs in r_state:
    print('Working on random state: {}'.format(rs))
    X_train, X_test, y_train, y_test, y_sc_train, y_sc_test = train_test_split(X, y, y_sc, 
                                                                                train_size=0.80, 
                                                                                random_state=rs)

    MLGP = OMGP(X_train=X_train,
                X_test=X_test,
                y_train=y_train,
                y_test=y_test,
                kernel_recipe=kernel_recipe,
                scaling=True, 
                scaling_params={'alpha' : sc.slope,
                                'gamma' : sc.intercept},
                scaling_y_train=y_sc_train,
                scaling_y_test=y_sc_test)

    MLGP.run_GP()
    r_data[rs] = MLGP.__dict__


if not os.path.exists('run_CH3'):
    os.mkdir('run_CH3')
        
a.savefig('run_CH3/scaling_plot.png')
np.save('run_CH3/scaling_data.npy', scaling_dict)

np.save('run_CH3/CH3_rsdata.npy', r_data)
PP = MLGP.parity_plot(data='train')
PP.savefig('run_CH3/CH3_parity_plot_train.png')
PP = MLGP.parity_plot(data='test')
PP.savefig('run_CH3/CH3_parity_plot_test.png')
PP = MLGP.parity_plot(data='train', err_bar=True)
PP.savefig('run_CH3/CH3_parity_plot_train_err_bar.png')
PP = MLGP.parity_plot(data='test', err_bar=True)
PP.savefig('run_CH3/CH3_parity_plot_test_err_bar.png')
```
